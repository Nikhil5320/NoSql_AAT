```python
import requests
import json
import matplotlib.pyplot as plt
```

```python
r = requests.get("http://localhost:6040/analytic1")

for i in r:
    dicty = json.loads(i)
print("Total Number of Records : ", dicty['one']+dicty['two'])
print("Total Number of people with heart disease : ", dicty['one'])
print("Total Number of healthy peole : ", dicty['two'])
print("Percentage of people with Heart Disease : ", dicty['one']/(dicty['one']+dicty['two'])*100, "%")
```

# Comparision of Male and Female Heart Disease Patients

```python
r1 = requests.get("http://localhost:6040/analytic3/male")
r2 = requests.get("http://localhost:6040/analytic4/female")
for i in r1:
    m = json.loads(i.decode())
for i in r2:
    f = json.loads(i.decode())
print(m,f)
x = ['M' for i in range(m)]+['M','F']+['F' for i in range(f)]
y = [i for i in range(m)]+[m,f]+[i for i in range(f)]
plt.plot(x, y)
plt.xlabel('Gender')
plt.ylabel('number of people with heart disease')
plt.title('Graph to compare male and female patients')
plt.show()
```

# Comparing the heart disease rate with different Chest Pain Type
